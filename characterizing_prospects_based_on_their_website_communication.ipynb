## Define a function that loop through the list of comp and grab row URL - returns a list of URLs

```python
def get_urls(companies):
    urls_final=[]
    i=0
    while i<len(companies):
        search_query = driver.find_element_by_name('q')
        search_query.clear()
        search_query.send_keys(companies[i])
        search_query.send_keys(Keys.ENTER)
        sleep(randint(2,6))
        address = driver.find_elements_by_class_name("TbwUpd")
        urls_final.append(address[0].text)
        sleep(randint(4,8))
        i=i+1
    return urls_final
urls_final = get_urls(companies)
```

## Define a function that cleans the grabbed URLS

```python
def clean_urls(urls_final):
    clean_urls = []
    for url in urls_final:
        url = url.split('â€º', 1)[0]
        url = url.split(' ', 1)[0]
        clean_urls.append(url)
    return clean_urls
clean_urls = clean_urls(urls_final)
```

## Check if the grab URL is consistant with the company name
compare the first letter after www. (the method is not optimal but allows to rise a warning if there is a chance to have grabbed the wrong URL)

```python
def checking_urls(companies, urls_final):
    '''A functions that return a warning if the first letter
    after www. is different from the first letter of the company name'''
    urls=urls_final
    j=0
    while j<len(urls):
        if (urls[j][0] !='w' and urls[j][1] !='w'):
            urls[j] = urls[j]
        else:
            try:
                urls[j] = urls[j].split('www.', 1)[1]
                #print(urls[j])
            except:
                urls[j]=urls[j]
        j=j+1
    i=0
    warning = []
    while i<len(companies):
        companies[i] = companies[i].lower()
        if companies[i][0] == urls[i][0]:
            warning.append('OK')
        else:
            warning.append('WARNING')
        i=i+1
    return warning
warnings = checking_urls(companies, urls_final)
```

# Looping through the companies name, entering the page, grabbing the HTML from Home page - checking if key words are present within and append to a list if it is the case or not

```python
def presence (companies, key_words):
    outer_presence_final = []
    i=0
    while i<len(companies):
        try:
            search_query = driver.find_element_by_name('q')
        except:
            driver.execute_script("window.history.go(-2)")
        search_query = driver.find_element_by_name('q')
        search_query.clear()
        search_query.send_keys(companies[i])
        sleep(randint(1,3))
        search_query.send_keys(Keys.ENTER)
        sleep(randint(1,2))
        link = driver.find_element_by_tag_name("cite")
        if '.' not in link.text:
            clickable = driver.find_elements_by_class_name('r')
            clickable[1].click()
            sleep(randint(4,6))
            page_html = driver.page_source
            page_soup = soup(page_html, 'html.parser')
            page_soup = soup(page_soup.text, 'html.parser') # Rajout
            page_soup = page_soup.text.replace('\n','') # Rajout
            page_soup = re.sub(r"(\w)([A-Z])", r"\1 \2", page_soup) # Rajout
            page_soup = re.sub(' +', ' ', page_soup)
            page_soup = page_soup.lower()
            page_soup = page_soup.replace(',','')
            page_soup = page_soup.replace('!','')
            page_soup = page_soup.replace(':','')
            j=0
            inner_presence = []
            while j<len(key_words):
                test = page_soup.findAll(text = re.compile(key_words[j]))
                if len(test)>0:
                    try:
                        inner_presence.append(test)
                    except:
                        inner_presence.append('fails')
                j=j+1
            driver.back()
            sleep(randint(2,4))
            outer_presence_final.append(inner_presence)
        else:
            link.click()
            sleep(randint(4,6))
            page_html = driver.page_source
            page_soup = soup(page_html, 'html.parser')
            page_soup = soup(page_soup.text, 'html.parser') # Rajout
            page_soup = page_soup.text.replace('\n','') # Rajout
            page_soup = re.sub(r"(\w)([A-Z])", r"\1 \2", page_soup) # Rajout
            page_soup = re.sub(' +', ' ', page_soup)
            page_soup = page_soup.lower()
            page_soup = page_soup.replace(',','')
            page_soup = page_soup.replace('!','')
            page_soup = page_soup.replace(':','')
            j=0
            inner_presence = []
            while j<len(key_words):
                test = page_soup.findAll(text = re.compile(key_words[j]))
                if len(test)>0:
                    try:
                        inner_presence.append(test)
                    except:
                        inner_presence.append('fails')
                j=j+1
            driver.back()
            sleep(randint(2,4))
            outer_presence_final.append(inner_presence)
        i=i+1
    return outer_presence_final
```
